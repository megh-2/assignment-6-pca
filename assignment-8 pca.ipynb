# Hierarchical cluster

```python
import scipy.cluster.hierarchy as sch  ## plotting dendrogram
from sklearn.cluster import AgglomerativeClustering  # to generate the clusters
import seaborn as sns
```

```python
#normalise function
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return x
```

```python
# NOrmalize data frame (comsidering the numerical part of data)
df_norm =norm_func(wine.iloc[:,1:])
```

```python
#create dendrogram
dendrogam = sch.dendrogram(sch.linkage(df_norm,method="complete"))
```

```python
#create clusters
hc  = AgglomerativeClustering(n_clusters=3,affinity  = "euclidean",linkage= "single")
```

```python
# save clusters fro chart
y_hc = hc.fit_predict(df_norm)
clusters = pd.DataFrame(y_hc,columns=["Clusters"])
```

```python
wine["h_clusterid"]=clusters
```

```python
wine.head()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(wine['h_clusterid'],wine['Type'], c=hc.labels_) 
```

# KMeans

```python
from sklearn.cluster import KMeans
```

```python
wine = pd.read_csv(r"C:\Users\91863\Downloads\wine.csv")
```

```python
# NOrmalisation function
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled_wine_df = scaler.fit_transform(wine.iloc[:,1:])
```

```python
wcss = []
for i in range(1,11):
    kmeans= KMeans(n_clusters = i,random_state=0)
    kmeans.fit(scaled_wine_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("NUMBER OF CLUSTERS")
plt.ylabel("WCSS")
plt.show()
```

the optimum number of clusters will be 3

```python
# BUild cluster algorithm
from sklearn.cluster import KMeans
cluster_new= KMeans(3,random_state=42)   # 42 is a random number 
cluster_new.fit(scaled_wine_df)
```

```python
cluster_new.labels_
```

```python
# assign clusters to the data set
wine["clusterid_new"]= cluster_new.labels_
```

```python
wine.head()
```

```python
# these are standard values
cluster_new.cluster_centers_
```

```python
wine.drop(columns="Type").groupby("clusterid_new").agg(["mean"]).reset_index()
```
